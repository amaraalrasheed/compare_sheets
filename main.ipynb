```python
import openpyxl

def compare_sheets(file_name, sheet1, sheet2):

    workbook = openpyxl.load_workbook(file_name)

    # Access the specified sheets
    sheet1 = workbook[sheet1]
    sheet2 = workbook[sheet2]

    # Compare the sheets and print differences
    for row in range(1, max(sheet1.max_row, sheet2.max_row) + 1):
        for col in range(1, max(sheet1.max_column, sheet2.max_column) + 1):
            cell1 = sheet1.cell(row, col)
            cell2 = sheet2.cell(row, col)

            if cell1.value != cell2.value:
                print(f"Difference found at cell {cell1.coordinate}:")
                print(f"  Sheet 1: {cell1.value}")
                print(f"  Sheet 2: {cell2.value}")

file_name = "master.xlsx"
sheet1 = "data"
sheet2 = "data_old"
compare_sheets(file_name, sheet1, sheet2)
```

```python
import openpyxl

def compare_sheets(file_name, sheet1, sheet2, col1, col2):

    workbook = openpyxl.load_workbook(file_name)

    # Access the specified sheets
    sheet1 = workbook[sheet1]
    sheet2 = workbook[sheet2]

    # Compare the specified columns and print differences
    for row in range(1, max(sheet1.max_row, sheet2.max_row) + 1):
        cell1 = sheet1.cell(row, col1)
        cell2 = sheet2.cell(row, col2)

        if cell1.value != cell2.value:
            print(f"Difference found at row {row}:")
            print(f"  Sheet 1: {cell1.value}")
            print(f"  Sheet 2: {cell2.value}")

file_name = "master.xlsx" # Put your excel file name here
sheet1 = "data" # Put your first sheet name here
sheet2 = "data_old" # Put your Second sheet name here
col1 = 5  # column A = 1 in  sheet1
col2 = 5  # column A = 1 in  sheet2
compare_sheets(file_name, sheet1, sheet2, col1, col2)
```
